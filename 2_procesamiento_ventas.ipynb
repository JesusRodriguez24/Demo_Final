```python
import pandas as pd #Se importa la libraria de panda 
```

```python
try:
    df_ventas = pd.read_csv("validados/ventas_clean.csv", index_col="Unnamed: 0", sep = ',', dtype={"UPC":  "string"})
except Exception as e:
       print("Error al momento de cargar el dataframe"+e)
finally:
        print("Se ha cargado el proceso de venta mensual")
#df_ventas =df_ventas.drop(columns=["Unnamed: 0"])
df_ventas.head(1000)
```

```python
#se carga el catalogo de tienda
df_pdv = pd.read_csv("validados/pdv_clean.csv",  index_col="Unnamed: 0", sep = ',',encoding = 'utf8')
#df_ventas =df_ventas.drop(columns=["Unnamed: 0"])
df_pdv.head(1000)
```

```python
#Se carga el catalogo de productos de las tiendas
df_sku = pd.read_csv("validados/productos_clean.csv",  index_col="Unnamed: 0", sep = ',',encoding = 'utf8', dtype={"UPC":  "string"})
df_sku.head(1000)
```

```python
#Procedemos a hacer el inner join inicialmente de los pdv
pd_datos_pdv= pd.merge(df_ventas, df_pdv, left_on ="tienda_id",  right_on ="ID", how = "inner" )
pd_datos_pdv
```

```python
#Se hace el join del catalogo de los sku
pd_datos_raw= pd.merge(pd_datos_pdv, df_sku, left_on ="Producto_id",  right_on ="ID", how = "inner" )
pd_datos_raw
```

```python
#Funcion para obtener la venta que seria la multiplicacio del precio por la cantidad
def get_venta_valor(precio, cantidad):
    return precio*cantidad
```

```python
#Se usa apply para crear un campo con la funcion  get_venta_valor que recibe dos parametros
pd_datos_raw['venta'] = pd_datos_raw.apply(lambda x: get_venta_valor(x['precio'], x['cantidad']), axis=1)
```

```python
#Se verifica los tipos de datos
pd_datos_raw.dtypes
```

```python
#Se verifica las dimensiones del nuevo datagrame
pd_datos_raw.shape
```

```python
#se verifica las cantidades para ver si el sku en cuanto pdv estan

pd_datos_raw.groupby("UPC")["Pos"].value_counts()
```

```python
#Se genera agrupacion de promedio mediana
venta_por_sku = pd_datos_raw.groupby('UPC')['venta'].agg(['mean', 'median', 'std'])
venta_por_sku
```

```python
#Se genera agrupacion de promedio mediana
venta_por_tienda = pd_datos_raw.groupby('Pos')['venta'].agg(['mean', 'median', 'std'])
venta_por_tienda
```

```python
#Se guarda los resultados en un csv
venta_por_sku.to_csv("resultados/venta_por_sku.csv")
```

```python
#Se guarda los resultados en un csv
venta_por_tienda.to_csv("resultados/venta_por_tienda.csv")
```
