```python
import pandas as pd #Se importa la libraria de panda 
import matplotlib.pyplot as plt
```

```python
try:
    venta_por_sku = pd.read_csv("resultados/venta_por_sku.csv", dtype={"UPC":  "string"})
    venta_por_sku.head(1000)
except Exception as e:
       print("Error al momento de cargar el dataframe"+e)
finally:
        print("Se ha cargado el proceso de venta mensual")
venta_por_sku
```

```python
try:
    venta_por_tienda = pd.read_csv("resultados/venta_por_tienda.csv")
    venta_por_tienda.head(1000)
except Exception as e:
       print("Error al momento de cargar el dataframe"+e)
finally:
        print("Se ha cargado el proceso de venta mensual")
venta_por_tienda
```

```python
#Se hace ordenamiento por rating de transaccion
#venta_por_tienda.set_index("Pos",inplace=True)
venta_por_tienda= venta_por_tienda.sort_values("mean", ascending=False)
venta_por_tienda
```

```python
#Se hace ordenamiento por rating de transaccion
#venta_por_sku.set_index("UPC",inplace=True)
venta_por_sku= venta_por_sku.sort_values("mean", ascending=False)
venta_por_sku
```

```python

%matplotlib inline
venta_por_tienda.plot(kind='bar', x='Pos', y='sum',color='green');
```

```python
%matplotlib inline
venta_por_sku.plot(kind='bar', x='UPC', y='sum',color='red');
```
