# Proyecto Final: Carga de datos desde csv de datos de supermercados

Para este demo se cargara la información desde tres archivos pdv
1. ventas_mensuales:  Archivos con las diferentes transacciones de un mes
2. pdv.csv: archivo maestro de pdv
3. sku.csv: archivo maestro de productos



```python
import pandas as pd #Se importa la libraria de panda 
```

# Carga Inicial de transacciones

```python
try:
    df_ventas = pd.read_csv("origen/ventas_mensuales.csv", sep = ';', dtype={"UPC":  "string"})
    df_ventas.head(1000)
except Exception as e:
       print("Error al momento de cargar el dataframe"+e)
finally:
        print("Se ha cargado el proceso de venta mensual")
```

```python
# usando df.shape() para obtener las dimensiones del dataframe generado
df_ventas.shape
```

```python
#Se inicializan los indices del dataframe
df_ventas.reset_index(drop=True)
```

```python
#Vericando si existen nan en cada columna
df_ventas.isna().sum()
```

```python
#Eliminando todos los nans del data frame pero generando una copia del dataframe para evitar 
#el problema de slice si es una vista o una copia

df_ventas_nonans= df_ventas.dropna().copy()
df_ventas_nonans
```

```python
#Se crea campo para poder hacer el ordernamiento o agrupacion adecuadamente
df_ventas_nonans['fecha_larga'] =  df_ventas_nonans["fecha"] +' '+   df_ventas_nonans["hora"]
df_ventas_nonans
```

```python
#Se guarda los resultados en un csv
df_ventas_nonans.to_csv("validados/ventas_clean.csv")
```
