```python
import pandas as pd #Se importa la libraria de panda 
```

```python
try:
    df_ventas = pd.read_csv("ventas_mensuales.csv", sep = ';', dtype={"UPC":  "string"})
    df_ventas.head(1000)
except Exception as e:
       print("Error al momento de cargar el dataframe"+e)
finally:
        print("Se ha cargado el proceso de venta mensual")
```

```python
# usando df.shape() para obtener las dimensiones del dataframe generado
df_ventas.shape
```

```python
#Se inicializan los indices del dataframe
df_ventas.reset_index(drop=True)
```

```python
#Vericando si existen nan en cada columna
df_ventas.isna().sum()
```

```python
#Eliminando todos los nans del data frame pero generando una copia del dataframe para evitar 
#el problema de slice si es una vista o una copia

df_ventas_nonans= df_ventas.dropna().copy()
df_ventas_nonans
```

```python
df_ventas_nonans
```

```python
#Se crea campo para poder hacer el ordernamiento o agrupacion adecuadamente
df_ventas_nonans['fecha_larga'] =  df_ventas_nonans["fecha"] +' '+   df_ventas_nonans["hora"]
```

```python
df_ventas_nonans
```

```python
#Se hace ordenamiento por rating de transaccion
df_ventas.sort_values("rating", ascending=False)
```

```python
#se carga el catalogo de tienda
df_pdv = pd.read_csv("pdv.csv", sep = ';',encoding = 'utf8')
df_pdv.head(1000)
```

```python
#Se carga el catalogo de productos de las tiendas
df_sku = pd.read_csv("sku.csv", sep = ';',encoding = 'utf8', dtype={"UPC":  "string"})
df_sku.head(1000)
```

```python
#Procedemos a hacer el inner join inicialmente de los pdv
pd_datos_pdv= pd.merge(df_ventas_nonans, df_pdv, left_on ="tienda_id",  right_on ="ID", how = "inner" )
pd_datos_pdv
```

```python
#Se hace el join del catalogo de los sku
pd_datos_raw= pd.merge(pd_datos_pdv, df_sku, left_on ="Producto_id",  right_on ="ID", how = "inner" )
pd_datos_raw
```

```python
def get_venta_valor(precio, cantidad):
    return precio*cantidad
```

```python
pd_datos_raw["cantidad"]
```

```python
pd_datos_raw.to_csv("prueba.csv")
```

```python
pd_datos_raw['venta'] = pd_datos_raw.apply(lambda x: get_venta_valor(x['precio'], x['cantidad']), axis=1)
```

```python
pd_datos_raw.groupby("UPC")["Pos"].value_counts()
```

```python
pd_datos_raw.groupby('UPC')['venta'].agg(['mean', 'median', 'std'])
```
